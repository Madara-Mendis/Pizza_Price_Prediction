```python
data.rename(columns={'price_rupiah': 'price'}, inplace=True)
```

```python
data['price']=data['price'].str.replace('Rp','')
data['price']=data['price'].str.replace(',','')
data['price']=data['price'].astype('int32')
```

```python
data.head()
```

```python
data.info()
```

```python
data['diameter']=data['diameter'].str.replace(' inch','').astype('float32')
```

```python
data.head()
```

```python
data[(data['size']=="jumbo")& (data['diameter']<=16)]
```

```python
data=data.drop(data.index[[6,11,16,80]])
```

```python
cat_cols= data.select_dtypes(include=['object']).columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
en=LabelEncoder()
for i in cat_cols:
    data[i]=en.fit_transform(data[i])
```

```python
data.head()
```

```python
X=data.drop('price',axis=1)
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
svm=SVR()
svm.fit(X_train,y_train)
```

```python
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
gbr=GradientBoostingRegressor()
gbr.fit(X_train,y_train)
```

```python
xg=XGBRegressor()
xg.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
y_pred2=svm.predict(X_test)
y_pred3=rf.predict(X_test)
y_pred4=gbr.predict(X_test)
y_pred5=xg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
score5=metrics.r2_score(y_test,y_pred5)
```

```python
print(score1)
print(score2)
print(score3)
print(score4)
print(score5)
```
